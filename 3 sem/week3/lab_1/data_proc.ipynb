```python
import matplotlib.pyplot as plt
import numpy as np

f1 = open('build/data/dice.txt')

prob1 = [float(i) for i in f1.readlines()]

# print(prob)

plt.hist([i for i in range(1,101)],100, weights=prob1)
plt.plot([1, 100],[1/100,1/100], color="red")

# plt.xlabel('Количество чисел в массиве')
# plt.ylabel('Время поиска')

# plt.plot(value, g(value))
```
